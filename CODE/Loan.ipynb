```python
#Cột LanguageCode
Data.loc[~Data['LanguageCode'].isin([1, 2, 3, 4, 5, 6, 9,19,22]), 'LanguageCode'] = 'Others'
language_code = {1: 'Estonian', 2: 'English', 3: 'Russian', 4: 'Finnish', 5: 'German', 6: 'Spanish', 9: 'Slovakian', 19: "Finland", 22:"Salvador"}
Data['LanguageCode'] = Data['LanguageCode'].replace(language_code)
```

```python
Data.Education.value_counts()
```

```python
# Cột Education
Education_list={1: 'Primary', 2: 'Basic', 3: 'Vocational',4:"Secondary", 5: "Higher", 0: "Unknown", -1: "Unknown" }
Data.Education= Data.Education.replace(Education_list)
```

```python
Data.MaritalStatus.value_counts()
```

```python
#Cột MarritalStatus
Marrital_list= {1:"Married", 2: "Cohabitant", 3: "Single", 4: "Divorced", 5: "Widow", 0:"NotSpecified", -1: "NotSpecified" }
Data.MaritalStatus= Data.MaritalStatus.replace(Marrital_list)
# Cột EmploymentStatus
Employment_list= {1: "UnEmployed", 2: "Partially Employed", 3:"Fully Employed", 4:"Self Employed", 5: "Entrepreneur", 6:"Retiree", -1: " Unknown", 0: "Unknown"  }
Data.EmploymentStatus= Data.EmploymentStatus.replace(Employment_list)
```

```python
Data.EmploymentDurationCurrentEmployer.value_counts()
```

```python
#Cột EmploymentDurationCurrentEmployer- Thay thế các giá trị null bằng Unknown
Data.EmploymentDurationCurrentEmployer= Data.EmploymentDurationCurrentEmployer.fillna("unknown")
# Cột HomeOwnershipType
HomeOwner_list= {0: "Homeless", 1: "Owner", 2:'LivingWithParents',3: 'TenantFurnished', 4:"TenantUnfurnished", 5:"CouncilTenant", 6:"JointTenant", 7: "JointOwner",
8:"OwnerMortgage", 9:'OwnerEncumbrance',10: "Other" ,-1:" Unknown"}
Data.HomeOwnershipType= Data.HomeOwnershipType.replace(HomeOwner_list)
```

```python
# Cột UseOfLoan
Use_list= {-1: "NotSet", 0:'Loan consolidation',1:'Real estate',2:'Home improvement',3:'Business',4:'Education', 5: 'Travel',6: 'Vehicle',
7: 'Other',8:'Health', 101:'Working capital financing',102:'Purchase of machinery equipment',104:'Accounts receivable financing',
106: 'Construction finance', 107:'Acquisition of stocks', 108:'Acquisition of real estate', 101:'Other business'}
Data.UseOfLoan= Data.UseOfLoan.replace(Use_list)
# Cột OccupationArea
Are_list= {1:"Other", 2:'Mining',3: 'Processing', 4:'Energy',5:'Utilities', 7: 'Retail_and_wholesale',6:'Construction', 8:'Transport_and_warehousing',
9: 'Hospitality and catering',10: 'Info and telecom',11:'Finance and insurance',12: 'Real estate',13:'Research',14:'Administrative',15:'Civil service and military',16:'Education',17:'Healthcare and social help',
18:'Art and entertainment',19:'Agriculture forestry and fishing',-1:'NotSpecified',0:'NotSpecified'}
Data.OccupationArea=Data.OccupationArea.replace(Are_list)
```

```python
Data.WorseLateCategory.value_counts()
```

```python
# Cột WorseLateCategory - Thay thế các giá trị null và giá trị 44933,45153 về giá trị 0-15
Worselist= {"44933": "0-15", "45153" : "0-15"}
Data.WorseLateCategory= Data.WorseLateCategory.replace(Worselist)
Data.WorseLateCategory.fillna("0-15", inplace= True)

```

```python
# Cột NewCreditCustomer
Data.loc[Data['NewCreditCustomer'] == False,'NewCreditCustomer'] = 'Existing Credit Customer'
Data.loc[Data['NewCreditCustomer'] == True,'NewCreditCustomer'] = 'New Credit Customer'
# Cột Restructured
Data.loc[Data['Restructured'] == False,'Restructured']='No'
Data.loc[Data['Restructured'] == True,'Restructured']='Yes'
# Cột ActiveScheduleFirstPaymentReache
Data.loc[Data['ActiveScheduleFirstPaymentReached'] == False,'ActiveScheduleFirstPaymentReached']='No'
Data.loc[Data['ActiveScheduleFirstPaymentReached'] == True,'ActiveScheduleFirstPaymentReached']='Yes'
```
