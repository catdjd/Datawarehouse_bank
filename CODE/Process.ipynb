```python
# Bảng Loan_Detail
Loan_Detail= Data[['LoanId','FirstPaymentDate','ActiveScheduleFirstPaymentReached',
       'LastPaymentOn', 'CurrentDebtDaysPrimary', 'DebtOccuredOn',
       'CurrentDebtDaysSecondary', 'DebtOccuredOnForSecondary','PrincipalOverdueBySchedule',
       'PrincipalPaymentsMade', 'InterestAndPenaltyPaymentsMade',
       'PrincipalWriteOffs', 'InterestAndPenaltyWriteOffs', 'PrincipalBalance',
       'InterestAndPenaltyBalance','GracePeriodStart',
       'GracePeriodEnd']]
Loan_Detail["Loan_detailID"]=range(1,len(Loan_Detail)+1)
Loan_Detail.isnull().sum()
```

Đánh giá: Những cột có giá trị NULL là những giá trị chứa các khoản đánh giá của các khoản vay bị trễ hạn và đang trong thời gian thu hồi nợ (tính đến ngày 22/10/2023- ngày cập nhập khoản vay) bao gồm thời gian nợ tiền gốc, tiền lãi suất, thời gian các khoản vay nợ( quá hạn không trả và người cho vay quyết định thu hồi nợ)
- Một số cột khác như gracePeriodSart hay End: thời gian gia hạn khoản vay chỉ áp dụng ở một số khoản vay có gia hạn
- Cột PrincipalWriteOffs và InterestAndPenaltyWriteOffs : chỉ áp dụng cho những khoản vay không có khả năng trả nợ và rơi vào vỡ nợ, người vay quyết định xóa nợ các khoản này
- Cột InterestAndPenaltyBalance: số tiền lãi và chi phí chưa thanh toán có thể NULL vì còn phụ thuộc vào khả năng trả nợ và các chi phí phát sinh liên quan sau này.
- Cột RecoveryStage: trạng thái thu hồi nợ đánh giá cho tất cả các khaorn vay, với các giá trị NULL chó thể thay bởi giá trị 0 vì chưa ở trạng thái phải thu hồi, người vay đang thanh toán đầy đủ khoản phí.

```python
# Xử lí giá trị NULL ở cột RecoveryStage, PrincipalOverdueBySchedule 

Loan_Detail.PrincipalOverdueBySchedule.fillna(0,inplace=True)
Loan_Detail.CurrentDebtDaysPrimary.fillna(0,inplace=True)
Loan_Detail.CurrentDebtDaysSecondary.fillna(0,inplace=True)
Loan_Detail.PrincipalWriteOffs.fillna(0,inplace=True)
Loan_Detail.InterestAndPenaltyWriteOffs.fillna(0,inplace=True)

#Loan_Detail.to_csv("Loan_Detail.csv", index=False)
Loan_Detail.head()
```
